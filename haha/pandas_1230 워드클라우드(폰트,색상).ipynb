### make_wordcsms는 colormap에 있는 색상으로 워드클라우드를 만들때 사용

### make_wordc(wc를 만들고자 하는 파일, masking할 이미지, 색상지정)

```python
def make_wordc(text, mask_image, color):
    wc = WordCloud(background_color="white", max_words=2000,
                  font_path='C:/Users/skato/AppData/Local/Microsoft/Windows/Fonts/Jalnan.ttf',
                  mask=mask_image, max_font_size=90,
                  colormap=color)
    
    wc = wc.generate_from_frequencies(dict(text))
    f = plt.figure(figsize=(12, 12))
    plt.imshow(wc, interpolation='bilinear')
    plt.axis('off')
```
